Import psycopg2 to connect to postgis database

```python
import psycopg2  # (if it is postgres/postgis)
```

```python
con = psycopg2.connect(database="thelab", user="postgres", password="postgres", host="localhost")
```

```python
sql = "select geom,id,floor_id from thelaboffice.ga_updated"

df = gpd.read_postgis(sql, con, geom_col='geom' )
```
